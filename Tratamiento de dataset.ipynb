```python
data = data.rename(columns={'name':'Nombre', 'pokedex_number':'Numero_pokedex', 'classfication':'Clasificacion', 'type1':'Primer_Tipo', 'type2':'Segundo_Tipo',
                            'weight_kg':'Peso', 'height_m':'Altura', 'percentage_male':'Probabilidad_Macho', 'abilities':'Habilidades',
                            'hp':'PS', 'attack':'Ataque', 'defense':'Defensa', 'sp_attack':'Ataque_Especial', 'sp_defense':'Defensa_Especial',
                            'speed':'Velocidad', 'base_total':'Total_Stats', 'against_bug':'Contra_Bicho', 'against_dark':'Contra_Siniestro', 'against_dragon':'Contra_Dragon',
                            'against_electric':'Contra_Electrico', 'against_fairy':'Contra_Hada', 'against_fight':'Contra_Lucha', 'against_fire':'Contra_Fuego',
                            'against_flying':'Contra_Volador', 'against_ghost':'Contra_Fantasma', 'against_grass':'Contra_Planta', 'against_ground':'Contra_Tierra',
                            'against_ice':'Contra_Hielo', 'against_normal':'Contra_Normal', 'against_poison':'Contra_Veneno', 'against_psychic':'Contra_Psiquico',
                            'against_rock':'Contra_Roca', 'against_steel':'Contra_Acero', 'against_water':'Contra_Agua', 'is_legendary':'Legendario', 'generation':'Generacion'
                           })
```

```python
nombre=data[(data['Numero_pokedex']>=0)]['Nombre']
num_poke=data[(data['Numero_pokedex']>=0)]['Numero_pokedex']
clasificacion=data[(data['Numero_pokedex']>=0)]['Clasificacion']
tipo1=data[(data['Numero_pokedex']>=0)]['Primer_Tipo']
tipo2=data[(data['Numero_pokedex']>=0)]['Segundo_Tipo']
peso=data[(data['Numero_pokedex']>=0)]['Peso']
altura=data[(data['Numero_pokedex']>=0)]['Altura']
Prop_macho=data[(data['Numero_pokedex']>=0)]['Probabilidad_Macho']
habilidad=data[(data['Numero_pokedex']>=0)]['Habilidades']
hp=data[(data['Numero_pokedex']>=0)]['PS']
attack=data[(data['Numero_pokedex']>=0)]['Ataque']
defense=data[(data['Numero_pokedex']>=0)]['Defensa']
sp_attack=data[(data['Numero_pokedex']>=0)]['Ataque_Especial']
sp_defense=data[(data['Numero_pokedex']>=0)]['Defensa_Especial']
speed=data[(data['Numero_pokedex']>=0)]['Velocidad']
base_total=data[(data['Numero_pokedex']>=0)]['Total_Stats']
bug=data[(data['Numero_pokedex']>=0)]['Contra_Bicho']
dark=data[(data['Numero_pokedex']>=0)]['Contra_Siniestro']
dragon=data[(data['Numero_pokedex']>=0)]['Contra_Dragon']
electric=data[(data['Numero_pokedex']>=0)]['Contra_Electrico']
fairy=data[(data['Numero_pokedex']>=0)]['Contra_Hada']
lucha=data[(data['Numero_pokedex']>=0)]['Contra_Lucha']
fuego=data[(data['Numero_pokedex']>=0)]['Contra_Fuego']
volador=data[(data['Numero_pokedex']>=0)]['Contra_Volador']
fantasma=data[(data['Numero_pokedex']>=0)]['Contra_Fantasma']
planta=data[(data['Numero_pokedex']>=0)]['Contra_Planta']
tierra=data[(data['Numero_pokedex']>=0)]['Contra_Tierra']
hielo=data[(data['Numero_pokedex']>=0)]['Contra_Hielo']
normal=data[(data['Numero_pokedex']>=0)]['Contra_Normal']
veneno=data[(data['Numero_pokedex']>=0)]['Contra_Veneno']
psiquico=data[(data['Numero_pokedex']>=0)]['Contra_Psiquico']
roca=data[(data['Numero_pokedex']>=0)]['Contra_Roca']
acero=data[(data['Numero_pokedex']>=0)]['Contra_Acero']
agua=data[(data['Numero_pokedex']>=0)]['Contra_Agua']
legendary=data[(data['Numero_pokedex']>=0)]['Legendario']
generation=data[(data['Numero_pokedex']>=0)]['Generacion']

pokedex=pd.DataFrame(nombre, columns=['Nombre'])
pokedex
```

```python
prop_hembra=100-data['Probabilidad_Macho']
prop_hembra
```

```python
pokedex=pokedex.assign(Numero_Pokedex=num_poke, Clasificacion=clasificacion, Tipo_1=tipo1, Tipo_2=tipo2, Peso=peso, Altura=altura, 
                      Probabilidad_Macho=Prop_macho, Probabilidad_Hembra=prop_hembra, Habilidades=habilidad, PS=hp, Ataque=attack, Defensa=defense, Ataque_Especial=sp_attack,
                      Defensa_Especial=sp_defense, Velocidad=speed, Total=base_total, Contra_Bicho=bug, Contra_Siniestro=dark, Contra_Dragon=dragon,
                      Contra_Electrico=electric, Contra_Hada=fairy, Contra_Lucha=lucha, Contra_Fuego=fuego, Contra_Volador=volador,
                      Contra_Fantasma=fantasma, Contra_Planta=planta, Contra_Tierra=tierra, Contra_Hielo=hielo, Contra_Normal=normal,
                      Contra_Veneno=veneno, Contra_Psiquico=psiquico, Contra_Roca=roca, Contra_Acero=acero, Contra_Agua=agua, Legendario=legendary,
                      Generacion=generation
                      
                     )
pokedex
```
